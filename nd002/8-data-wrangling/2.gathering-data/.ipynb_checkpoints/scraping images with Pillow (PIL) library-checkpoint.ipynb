# Why use PIL and not just requests

Sometimes using requests, it can be error prone when downloading images and binary files. Therefore the Pillow library and the bytes IO library is best used for non text requests like images

```python
import requests
from PIL import Image
from io import BytesIO
```

```python
img_url = 'https://www.google.co.uk/images/branding/googlelogo/2x/googlelogo_color_272x92dp.png'
img = requests.get(img_url)
img_open = Image.open(BytesIO(img.content))
img_open
```
