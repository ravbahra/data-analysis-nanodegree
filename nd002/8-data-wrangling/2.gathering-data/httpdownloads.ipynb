```python
# create a directory to save all the files

# download directory
download_folder = 'download_folder'
download_path = '.\\' + download_folder
# first check if the directory exists
if (not os.path.exists('.\\' + download_folder)):
    os.mkdir(download_folder)
```

```python
# function will accept an url as an argument and write the file out

def http_download_and_save(link):
    # get the text from the file
    text = requests.get(link).text #can use .content too
    # get the filename from the link
    filename = link.split('/')[-1]
    # could also use the string.join()
    filepath = download_path + '\\' + filename

    # write the file, open is a context manager, meaning its object can be used using 'with' statement
    
    # os.path.join(folder, filename)
    with open(filepath, 'w') as downloaded_file:
        downloaded_file.write(text)
```

```python
# loop through all the files and download them

for download_url in ebert_review_urls:
    http_download_and_save(download_url)
```

## Reading files using glob

Now that we've downloaded all the files, we need to read them all in, we can use the glob library which allows you to use patterns for pathnames

```python
import glob
import pandas as pd
```

```python
# empty list to 
df_list = []

for ebert_review in glob.glob(download_path + '\\*.txt'):
    # open the file using utf8 encoding
    with open(ebert_review, 'r', encoding='utf8') as file:
        # first line is the title, will include the \n so do from start to last but one character
        title = file.readline()[:-1]
        # second line is the link
        review_url = file.readline()[:-1]
        
        # now read the rest of the text file
        review_text = file.read()
        
        # put all of these in a dictionary, which we'll append to a list to create a dataframe later
        df_list.append({
            'title':title,
            'review_url':review_url,
            'review_text':review_text
        })
```

```python
# specify columns so we have the order 
df = pd.DataFrame(df_list, columns=['title', 'review_url', 'review_text'])
```
